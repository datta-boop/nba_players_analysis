```python
import pandas as pd
```

```python
file_path='all_seasons_cleaned.csv'
data=pd.read_csv(file_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Top 10 Colleges by Number of Players

```python
top_colleges = data['college'].value_counts().head(10).reset_index()
top_colleges.columns = ['college','player_count']
```

# Bar Plot for Top Colleges

```python
plt.figure(figsize=(12,6))
sns.barplot(data=top_colleges, x='player_count', y='college', palette='coolwarm')
plt.title('Top 10 Colleges Producing Players')
plt.xlabel('Number of Players')
plt.ylabel('College')
plt.show()
```

# Average Performance Metrics by College( Top 5 Colleges with the Most Players)

```python
top_college_names = top_colleges['college']
college_performance = data[data['college'].isin(top_college_names)] \
    .groupby('college')[['pts', 'reb', 'ast']].mean().reset_index()
```

# Bar Plot for Average performance metrics by College

```python
college_performance.set_index('college').plot(kind='bar', figsize=(12, 6), color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Average Points, Rebounds, and Assists by Top Colleges')
plt.xlabel('College')
plt.ylabel('Average Metric Value')
plt.xticks(rotation=45)
plt.legend(title='Metrics')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Distribution of Points by College(Top 5 Colleges)

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[data['college'].isin(top_college_names)], x='college', y='pts', palette='muted')
plt.title('Distribution of Points by Top Colleges')
plt.xlabel('College')
plt.ylabel('Points')
plt.xticks(rotation=45)
plt.show()
```
