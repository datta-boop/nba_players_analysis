```python
import pandas as pd
```

```python
file_path='all_seasons.csv'
data=pd.read_csv(file_path)
```

```python
print("Initial Dataset Information:")
print(data.info())
```

```python
print("\nMissing value:\n",data.isnull().sum())
```

```python
print("\nDuplicate Rows\n:",data.duplicated().sum())
```

```python
print("\nSample Data:\n",data.head())
```

## There is no Unnamed column and no missing values.

```python
if 'Unnamed:0' in data.columns:
    data.drop(columns=['Unnamed:0'], inplace=True)
```

## Convertting the height from cm to meters for better readability

```python
data['player_height'] = data['player_height']/100
```

# Validate numeric ranges to check for anomalies

```python
invalid_heights=data[data['player_height']>2.5]
invalid_weights = data[data['player_weight']>200]
invalid_ages= data[data['age']<18]
```

# Output anomalies if found 

```python
print("\nInvalid Heights:\n", invalid_heights)
print("\nInvalid Weights:\n", invalid_weights)
print("\nInvalid Age:\n", invalid_ages)
```

## Display the Cleaned data

```python
print("\nCleaned Data Sample:\n", data.head())
```

Renaming the column

```python
data.rename(columns={'pts':'Points'},inplace=True)
data.head()
```

Grouping the points for each player

```python
import matplotlib.pyplot as plt
```

```python
player_performance = data.groupby('player_name')['Points'].sum().reset_index()
player_performance = player_performance.sort_values(by='Points', ascending=False)
top_10_players = player_performance.head(10).reset_index(drop=True)
```

Plotting the chart

```python
plt.figure(figsize=(10, 6))
plt.bar(top_10_players['player_name'], top_10_players['Points'], color=plt.cm.Reds(range(0, 256, 25)))
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Players by Points', fontsize=14)
plt.xlabel('Player Name', fontsize=12)
plt.ylabel('Total Points', fontsize=12)
plt.tight_layout()  
plt.show()
```

```python

import seaborn as sns
team_performance = data.groupby('team_abbreviation')['Points'].sum().reset_index()
team_performance = team_performance.sort_values(by='Points', ascending=False)
top_10_teams = team_performance.head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x='Points', y='team_abbreviation', data=top_10_teams,hue='team_abbreviation', palette='Reds')
plt.title('Top 10 Teams by Total Points', fontsize=16)
plt.xlabel('Total Points', fontsize=12)
plt.ylabel('Team', fontsize=12)


for index, value in enumerate(top_10_teams['Points']):
    plt.text(value + 1, index, str(round(value, 2)), va='center', fontsize=10)


plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans


player_performance = data.groupby('player_name')['Points'].sum().reset_index()
top_10_players = player_performance.sort_values(by='Points', ascending=False).head(10)
top_10_player_metrics = data[data['player_name'].isin(top_10_players['player_name'])]


performance_metrics = top_10_player_metrics[['Points', 'reb', 'ast']] 

kmeans = KMeans(n_clusters=3, random_state=0)
top_10_player_metrics['Cluster'] = kmeans.fit_predict(performance_metrics)


plt.figure(figsize=(10, 6))


sns.scatterplot(x=top_10_player_metrics['Points'], 
                y=top_10_player_metrics['reb'], 
                hue=top_10_player_metrics['Cluster'], 
                palette='Set1', s=100, marker='o', edgecolor='black')


plt.title('Top 10 Performers Clustered by Points and Rebounds')
plt.xlabel('Points')
plt.ylabel('Rebounds')


plt.show()
```

```python
#age wise performance 
```

```python
player_performance = data.groupby('player_name')['Points'].sum().reset_index()
top_30_players = player_performance.sort_values(by='Points', ascending=False).head(20)
top_30_player_data = data[data['player_name'].isin(top_30_players['player_name'])]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='Points', data=top_30_player_data, hue='Points', 
                palette=sns.light_palette("red", as_cmap=True), s=100, edgecolor='black')
plt.title('Player Age vs Points (Top 30)', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Points', fontsize=14)
plt.show()
```

```python
HoU Team performance over season
```

```python
team_performance_season = data.groupby(['team_abbreviation', 'season'])['Points'].sum().reset_index()
hou_performance = team_performance_season[team_performance_season['team_abbreviation'] == 'HOU']

plt.figure(figsize=(10, 6))
sns.lineplot(x='season', y='Points', data=hou_performance, marker='o')
plt.title('HOU Team Performance Over Seasons')
plt.xlabel('Season')
plt.ylabel('Total Points')
plt.xticks(rotation=45)
plt.show()
```
