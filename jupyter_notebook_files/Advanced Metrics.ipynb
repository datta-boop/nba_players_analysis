```python
import pandas as pd
```

```python
file_path='all_seasons_cleaned.csv'
data=pd.read_csv(file_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#  Distribution of Advanced Metrics

```python
sns.set_theme(style="whitegrid", context="notebook", palette="deep")
plt.figure(figsize=(12, 6))
sns.kdeplot(data=data, x='usg_pct', fill=True, color='#1f77b4', alpha=0.6, linewidth=2, label='Usage Rate (usg_pct)')
sns.kdeplot(data=data, x='ts_pct', fill=True, color='#2ca02c', alpha=0.6, linewidth=2, label='True Shooting % (ts_pct)')
sns.kdeplot(data=data, x='net_rating', fill=True, color='#d62728', alpha=0.6, linewidth=2, label='Net Rating')
plt.title('Distribution of Advanced Metrics', fontsize=16, fontweight='bold')
plt.xlabel('Metric Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(title='Metrics', fontsize=10)
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Top 10 Players by Net Rating
```

```python
 top_net_rating = data.nlargest(10, 'net_rating')[['player_name', 'net_rating']]

plt.figure(figsize=(12, 6))
sns.barplot(data=top_net_rating, x='net_rating', y='player_name', palette='Reds_r', edgecolor='black')
plt.title('Top 10 Players by Net Rating', fontsize=16, fontweight='bold')
plt.xlabel('Net Rating', fontsize=12)
plt.ylabel('Player Name', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data[['usg_pct', 'ts_pct', 'net_rating']].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=1.5, linecolor='black', cbar_kws={'shrink': 0.8})
plt.title('Correlation Between Advanced Metrics', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
