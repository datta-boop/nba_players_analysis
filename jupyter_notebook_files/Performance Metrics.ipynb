```python
import pandas as pd
```

```python
file_path='all_seasons_cleaned.csv'
data=pd.read_csv(file_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Top 10 Players by Points, Rebounds and Assists
top_points = data.nlargest(10,'pts')[['player_name','pts']]
top_rebounds=data.nlargest(10,'reb')[['player_name','reb']]
top_assists = data.nlargest(10,'ast')[['player_name','ast']]

# Leaderboard for Points
plt.figure(figsize=(12,6))
sns.barplot(data = top_points, x ='pts' , y='player_name' , palette='Blues_d')
plt.title('Top 10 Players by Points')
plt.xlabel('Points')
plt.ylabel('Player Name')
plt.show()
```

```python
## Leaderboard for Rebound
plt.figure(figsize=(12,6))
sns.barplot(data=top_rebounds, x='reb', y='player_name', palette='Greens_d')
plt.title('Top 10 Players by Rebounds')
plt.xlabel('Rebounds')
plt.ylabel('Player Name')
plt.show()
```

```python
## Leaderboard for Assists

plt.figure(figsize=(12,6))
sns.barplot(data=top_assists, x='ast', y='player_name', palette='Reds_d')
plt.title('Top 10 Players by Assists')
plt.xlabel('Assists')
plt.ylabel('Player Name')
plt.show()
```

```python
## Distributiobn of Points, Rebounds and Assists

plt.figure(figsize=(10,6))
sns.boxplot(data=data[['pts','reb','ast']],palette='pastel')
plt.title("Distribution of Points, Rebounds and Assists")
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
```
